```python
import pandas as pd 
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import r2_score, f1_score, recall_score, precision_score, accuracy_score

print('yea')
```

```python
df = pd.read_csv(r"C:\Users\abhis\jupyter and DS\Loan ML practice\LoanDataset - LoansDatasest.csv")
df
```

```python
df.dtypes
```

```python
df.isnull().sum().sort_values()
```

```python
df['loan_amnt'].fillna(df['loan_amnt'].mode()[0], inplace=True)  # Mode for loan_amnt
df['Current_loan_status'].fillna(df['Current_loan_status'].mode()[0], inplace=True)  # Mode for loan_amnt
df['loan_int_rate'].fillna(df['loan_int_rate'].mean(), inplace=True)  # Mean for loan_int_rate
df['employment_duration'].fillna(df['employment_duration'].median(), inplace=True)  # Median for employment_duration

df.dropna(subset=['customer_id'], inplace=True)
```

```python
df.isna().sum().sort_values()
```

```python
cols = ['loan_amnt', 'customer_income', 'employment_duration']

for col in cols:
    df[col] = df[col].astype(str).str.replace('£', '').str.replace('.00', '').str.replace('.0', '').str.replace(',', '').astype(int)
print(':)')
```

```python
df.head(2)
```

```python
df['loan_amnt'].value_counts()
```

```python
le = LabelEncoder()
df['loan_intent'] = le.fit_transform(df['loan_intent'])
df['home_ownership'] = le.fit_transform(df['home_ownership'])
df['loan_grade'] = le.fit_transform(df['loan_grade'])
df['cred_hist_length'] = le.fit_transform(df['cred_hist_length'])
```

```python
df.isnull().sum().sort_values()
```

```python
x = df.drop(columns=['Current_loan_status','customer_id'])
y = df['Current_loan_status']
```

```python
xtrain, xtet, ytrain, ytet = train_test_split(x,y,test_size = 0.25, random_state = 42)
```

# Without hyperparameter tuning

```python
rfc = RandomForestClassifier()
rfc.fit(xtrain,ytrain)
print(':)')
```

```python
y_pred = rfc.predict(xtet)
```

```python
y_pred
```

```python
# Model Evaluation Scores
accuracy = accuracy_score(ytet, y_pred)
f1 = f1_score(ytet, y_pred, average='weighted')  # F1 score
recall = recall_score(ytet, y_pred, average='weighted')  # Recall
precision = precision_score(ytet, y_pred, average='weighted')  # Precision

print(f"Accuracy score : {accuracy}")
print(f"F1 Score: {f1}")
print(f"Recall Score: {recall}")
print(f"Precision Score: {precision}")
```

# With hyperparameter tuning

```python
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}
```

```python
rfc_tuned = RandomForestClassifier(random_state=42)
grid_search = GridSearchCV(estimator=rfc_tuned, param_grid=param_grid, cv=3, verbose=2)
grid_search.fit(xtrain, ytrain)
```

```python
# Best parameters found by GridSearchCV
print("\nBest parameters found by GridSearchCV:")
print(grid_search.best_params_)

```

```python
# Predictions using the best model from GridSearchCV
best_rf_model = grid_search.best_estimator_
y_pred_tuned = best_rf_model.predict(xtet)
accuracy_tuned = accuracy_score(ytet, y_pred_tuned)  # R^2 score (not applicable for classification)
f1_tuned = f1_score(ytet, y_pred_tuned, average='weighted')  # F1 score
recall_tuned = recall_score(ytet, y_pred_tuned, average='weighted')  # Recall
precision_tuned = precision_score(ytet, y_pred_tuned, average='weighted')  # Precision

print(f"Accuracy Score: {accuracy_tuned}")
print(f"F1 Score: {f1_tuned}")
print(f"Recall Score: {recall_tuned}")
print(f"Precision Score: {precision_tuned}")
```
